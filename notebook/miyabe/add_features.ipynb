```python
TRAIN_DATA_PATH = '../train.csv'
TEST_DATA_PATH = '../test.csv'
```

```python
train = pd.read_csv(TRAIN_DATA_PATH)
test = pd.read_csv(TEST_DATA_PATH)
```
